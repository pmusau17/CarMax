```python
# Load relevant packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as sm
import warnings
from sklearn.cluster import KMeans 
```

```python
carmax_df = pd.read_csv('CaseDataRemodeled.csv')
```

```python
import requests
from bs4 import BeautifulSoup as bsoup
    
#assign variable for url from which we acquire html data
search_url = 'https://en.wikipedia.org/wiki/Automotive_industry'
                
#use requests pack to pull dat. Headers makes it think browser and not programatic
response = requests.get(search_url, headers={
            "User-Agent": "Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36"}).content
            
""" parse html """
parsed_html = bsoup(response, 'lxml')
""" initialize results list """
my_result_list = []

#for loop to dig into the table and create a list of lists
for result in parsed_html.find_all('tr')[64:133]:
    results1 = result.find_all('td')
    temp = []
    for i in results1:
        temp.append(i.text.strip().upper())
    try:
        my_result_list.append([temp[0],temp[1]])
    except:
        continue
```

```python
manualMake = [['BMW', 'GERMANY'],
             ['HUMMER', 'UNITED STATES'],
             ['ISUZU', 'JAPAN'],
              ['JAGUAR', 'UNITED KINGDOM'],
              ['LAND ROVER', 'UNITED KINGDOM'],
              ['LOTUS', 'UNITED KINGDOM'],
              ['MAZDA', 'JAPAN'],
              ['MERCEDES-BENZ', 'GERMANY'],
              ['MERCURY', 'UNITED STATES'],
              ['MINI', 'UNITED KINGDOM'],
              ['MITSUBISHI', 'JAPAN'],
              ['OLDSMOBILE', 'UNITED STATES'],
              ['PONTIAC', 'UNITED STATES'],
              ['SAAB', 'SWEDEN'],
              ['SATURN', 'UNITED STATES'],
              ['SCION', 'UNITED STATES'],
              ['SMART', 'SWITZERLAND'],
              ['SUBARU', 'JAPAN'],
              ['SUZUKI', 'JAPAN'],
              ['VOLVO', 'SWEDEN']]
```

```python
for i in manualMake:
    my_result_list.append(i)
```

```python
made_in = []
for i in range(len(list(carmax_df['purchase_make']))):
    carMade = 'NF'
    for manufacturer, country in my_result_list:
        if carmax_df['purchase_make'][i] == manufacturer:
                carMade = country
                break
        else:
            continue
    
    made_in.append(carMade)
```

```python
np.unique(np.array(made_in))
```

```python
carmax_df['made_in'] = made_in
```
